Dataset Loading and Inspection 

```python
import pandas as pd

df = pd.read_csv('../data/ipl_ball_by_ball_delivery.csv')

df.head()
df.columns
df.info()
```

Total Runs by Team

```python
df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
```

Run Distribution By Overs(runs mentioned per ball)

```python
df.groupby('over')['total_runs'].mean()
```

Chasing VS Batting First

```python
df.groupby('is_chasing')['total_runs'].mean()
```

Pressure Analysis

```python
df[df['is_chasing'] == 1].groupby('required_run_rate')['total_runs'].mean()
```
